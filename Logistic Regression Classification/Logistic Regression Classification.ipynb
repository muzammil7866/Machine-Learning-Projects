```python
import pandas as pd

train_data = {'x0': [1,1,1,1,1,1,1,1],'x1': [60,62,67,70,71,72,75,78], 'x2': [22,25,24,20,15,14,14,11], "y": [0,0,0,0,1,1,1,1]}
test_data = {'x0': [1,1,1,1], 'x1': [61,71,73,79],'x2': [23,19,15,13], "y": [0,0,1,1]}

train_data = pd.DataFrame(train_data)
test_data = pd.DataFrame(test_data)

#train_data[["x0", "x1", "x2"]] = (train_data[["x0", "x1", "x2"]] - train_data[["x0", "x1", "x2"]].mean()) / train_data[["x0", "x1", "x2"]].std()

theta0 = 0.1
theta1 = 0.1
theta2 = 0.1

m1 = train_data.shape[0]
m2 = test_data.shape[0]

print(train_data)





```

```python
component1 = []
for i in range(m1):
    component1.append(train_data.iloc[i]['x0']*theta0)

component2 = []
for i in range(m1):
    component2.append(train_data.iloc[i]['x1']*theta1)

component3 = []
for i in range(m1):
    component3.append(train_data.iloc[i]['x2']*theta2)



```

```python
import numpy as np

def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python

predictions = []
for i in range(m1):
     predictions.append(component1[i] + component2[i] + component3[i])

train_data['predictions'] = predictions

train_data['probabilities'] = sigmoid(train_data['predictions'])

train_data['h(x)-y'] = train_data['probabilities'] - train_data['y']

train_data
```

```python

#cost function

train_data['errors'] = train_data['y']*np.log(train_data['probabilities']) + (1 - train_data['y'])*np.log(1 - train_data['probabilities'])

cost = (-1/m1)*train_data['errors'].sum()
print("Cost is:", cost)
print(train_data)
```

```python
def generatePredictionAndError(train_data, theta0, theta1, theta2):

    train_data['predictions'] = theta0 * train_data['x0'] + theta1 * train_data['x1'] + theta2 * train_data['x2']
    train_data['probabilities'] = sigmoid(train_data['predictions'])

    train_data['h(x)-y'] = train_data['probabilities'] - train_data['y']

    return train_data
```

```python
#Mean Absolute Error (MAE)


originalDF = train_data
learningrate = 0.01
#learningrate = 0.001
#learningrate = 0.1

print("ORIGINAL THETA 0", theta0)
print("ORIGINAL THETA 1", theta1)
print("ORIGINAL THETA 2", theta2)


for i in range(100):
    
    theta0 = theta0 - (learningrate/m1)*((np.sign(train_data['h(x)-y'])*train_data['x0']).sum()) #gradient
    theta1 = theta1 - (learningrate/m1)*((np.sign(train_data['h(x)-y'])*train_data['x1']).sum())
    theta2 = theta2 - (learningrate/m1)*((np.sign(train_data['h(x)-y'])*train_data['x2']).sum())
    
    train_data = generatePredictionAndError(train_data, theta0, theta1, theta2)

print("UPDATED THETA 0", theta0)
print("UPDATED THETA 1", theta1)
print("UPDATED THETA 2", theta2)
print(train_data)
```

```python
final = []

for i in range(m1):
    if(train_data['probabilities'].iloc[i] >= 0.5):
        final.append(1)

    else:
        final.append(0)

train_data['final'] = final
print(train_data)
```

```python
print(test_data)

test_data = generatePredictionAndError(test_data, theta0, theta1, theta2)

final = []

for i in range(m2):
    if(test_data['probabilities'].iloc[i] >= 0.5):
        final.append(1)

    else:
        final.append(0)

test_data['final'] = final
print(test_data)
```

```python
#works best for 0.01
#not works as good for 0.001
#not works as good for 0.1
```
