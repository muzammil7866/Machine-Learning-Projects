```python
from sklearn.preprocessing import StandardScaler

bmi = features[['BMI']]
PhysicalHealth = features[['PhysicalHealth']]
MentalHealth = features[['MentalHealth']]
SleepTime = features[['SleepTime']]

scaler = StandardScaler()

bmi = scaler.fit_transform(bmi)
PhysicalHealth = scaler.fit_transform(PhysicalHealth)
MentalHealth = scaler.fit_transform(MentalHealth)
SleepTime = scaler.fit_transform(SleepTime)

data1_cleaned
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

x = features.copy()
test = x.copy()

test['BMI'] = bmi
test['PhysicalHealth'] = PhysicalHealth
test['MentalHealth'] = MentalHealth
test['SleepTime'] = SleepTime

le = LabelEncoder()
test = test.apply(le.fit_transform)

# Generate the correlation matrix
correlation_matrix = test.corr()

# Plot the correlation matrix heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix (Kidney Disease Dataset)')
plt.show()

# List of unnecessary features to drop
features_to_drop = ['HeartDisease', 'Race', 'Sex', 'SleepTime']

# Dropping the unnecessary columns from the dataframe
x = x.drop(columns=features_to_drop)

print(x)
x = pd.get_dummies(x, drop_first=True)

print(x)
encoded_df = x.apply(le.fit_transform)


encoded_df['BMI'] = bmi
encoded_df['PhysicalHealth'] = PhysicalHealth
encoded_df['MentalHealth'] = MentalHealth

print(encoded_df.columns)
encoded_df

```

# SUPPORT VECTOR MACHINE (SVM CLASSIFIER) 

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder, StandardScaler

x_train, x_test, y_train, y_test = train_test_split(
    encoded_df, label_encoded, test_size=0.8, stratify=label_encoded, random_state=42
)

# Instantiate and train the SVM model with default parameters
svm_model = SVC(probability=True, kernel='rbf', C=1, gamma='scale', random_state=42)
svm_model.fit(x_train, y_train)

# Predictions on the test set
svm_predictions = svm_model.predict(x_test)

# Accuracy
svm_accuracy = accuracy_score(y_test, svm_predictions)
print("Accuracy of SVM model is ", svm_accuracy * 100, "%")
print("Classification Report of SVM model is \n", classification_report(y_test, svm_predictions))

# Confusion Matrix
cm_svm = confusion_matrix(y_test, svm_predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix Heatmap (SVM)')
plt.show()

# ROC Curve (for binary classification)
if len(le.classes_) == 2:  # Only applicable for binary classification
    y_prob_svm = svm_model.predict_proba(x_test)[:, 1]
    fpr_svm, tpr_svm, _ = roc_curve(y_test, y_prob_svm)
    auc_score_svm = roc_auc_score(y_test, y_prob_svm)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr_svm, tpr_svm, color='green', label=f'AUC = {auc_score_svm:.2f}')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve (SVM)')
    plt.legend()
    plt.show()

# Feature Correlation Matrix Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(encoded_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix Heatmap')
plt.show()

# Ensure label_encoded is a Series
if isinstance(label_encoded, pd.DataFrame):
    label_encoded = label_encoded.iloc[:, 0]  # Extract the first column (adjust if needed)

# Class Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=label_encoded)
plt.title('Class Distribution')
plt.xlabel('Target Variable')
plt.ylabel('Count')
plt.show()


# SVM with Regularization
regularization_values = [0.1, 1, 10, 100, 1000]  # Different values for the regularization parameter C
svm_accuracies = []

for c_value in regularization_values:
    # Train SVM model with varying C values
    svm_model_reg = SVC(probability=True, kernel='rbf', C=c_value, gamma='scale', random_state=42)
    svm_model_reg.fit(x_train, y_train)
    
    # Predict and calculate accuracy
    svm_predictions_reg = svm_model_reg.predict(x_test)
    acc_reg = accuracy_score(y_test, svm_predictions_reg)
    svm_accuracies.append(acc_reg)

# Plot Effect of L2 Regularization on Accuracy
plt.figure(figsize=(8, 6))
plt.plot(regularization_values, svm_accuracies, marker='o', color='blue', label='SVM Accuracy')
plt.xscale('log')  # Log scale for better visualization
plt.xlabel('Regularization Strength (C)')
plt.ylabel('Accuracy')
plt.title('Effect of L2 Regularization on Accuracy (SVM)')
plt.legend()
plt.grid(True)
plt.show()

# Confusion Matrix for the Best Regularization (C=1)
best_c = 1  # Using C=1 as the best value (adjust based on results)
svm_model_best = SVC(probability=True, kernel='rbf', C=best_c, gamma='scale', random_state=42)
svm_model_best.fit(x_train, y_train)
best_svm_predictions = svm_model_best.predict(x_test)

cm_best_svm = confusion_matrix(y_test, best_svm_predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_best_svm, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title(f'Confusion Matrix Heatmap (SVM with C={best_c})')
plt.show()

```

### stratify=label_encoded ensures that the target labels are distributed equally between the train and test sets.

### SVC(kernel='rbf'): This initializes the Support Vector Classifier with an RBF (Radial Basis Function) kernel, which is commonly used for non-linear classification problems.

### C=1 is the regularization parameter that controls the trade-off between achieving a low error on the training data and a low model complexity (bias-variance trade-off).

### gamma='scale' is used to scale the kernel parameter for better performance.

### classification_report generates metrics such as precision, recall, and F1-score.

###  sns.heatmap to visualize the performance of the model.

### roc_curve and roc_auc_score are used for binary classification to evaluate the trade-off between true positive rate (TPR) and false positive rate (FPR).

### The area under the curve (AUC) gives a measure of how well the model distinguishes between classes.

### Regularization helps control overfitting by adding a penalty for large coefficients.
### The parameter C controls the strength of regularization: a smaller C leads to more regularization, while a larger C allows for less regularization.

### The feature correlation matrix is plotted to understand the relationships between features in the dataset.
### Strong correlations (positive or negative) might suggest the need for feature engineering or removal of redundant features.



# Random Forsest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# Instantiate and train the Random Forest model
rf_model = RandomForestClassifier(n_estimators=150, random_state=42)
rf_model.fit(x_train, y_train)

# Predictions on the test set
rf_predictions = rf_model.predict(x_test)

# Accuracy
rf_accuracy = accuracy_score(y_test, rf_predictions)
print("Accuracy of Random Forest model is ", rf_accuracy * 100, "%")
print("Classification Report of Random Forest model is \n", classification_report(y_test, rf_predictions))

# Confusion Matrix
cm_rf = confusion_matrix(y_test, rf_predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix Heatmap (Random Forest)')
plt.show()

# ROC Curve (for binary classification)
if len(le.classes_) == 2:  # Only applicable for binary classification
    y_prob_rf = rf_model.predict_proba(x_test)[:, 1]
    fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
    auc_score_rf = roc_auc_score(y_test, y_prob_rf)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr_rf, tpr_rf, color='orange', label=f'AUC = {auc_score_rf:.2f}')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve (Random Forest)')
    plt.legend()
    plt.show()

# Feature Correlation Matrix Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(encoded_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix Heatmap')
plt.show()

# Ensure label_encoded is a Series
if isinstance(label_encoded, pd.DataFrame):
    label_encoded = label_encoded.iloc[:, 0]  # Extract the first column (adjust if needed)

# Class Distribution
plt.figure(figsize=(8, 6))
sns.countplot(x=label_encoded)
plt.title('Class Distribution')
plt.xlabel('Target Variable')
plt.ylabel('Count')
plt.show()

# Random Forest with Different Estimators
n_estimators_values = [50, 100, 150, 200, 250]
rf_accuracies = []

for n_estimators in n_estimators_values:
    # Train Random Forest model with varying number of trees
    rf_model_varied = RandomForestClassifier(n_estimators=n_estimators, random_state=42)
    rf_model_varied.fit(x_train, y_train)
    
    # Predict and calculate accuracy
    rf_predictions_varied = rf_model_varied.predict(x_test)
    acc_varied = accuracy_score(y_test, rf_predictions_varied)
    rf_accuracies.append(acc_varied)

# Plot Effect of Number of Estimators on Accuracy
plt.figure(figsize=(8, 6))
plt.plot(n_estimators_values, rf_accuracies, marker='o', color='purple', label='RF Accuracy')
plt.xlabel('Number of Estimators')
plt.ylabel('Accuracy')
plt.title('Effect of Number of Estimators on Accuracy (Random Forest)')
plt.legend()
plt.grid(True)
plt.show()

# Confusion Matrix for the Best Number of Estimators (n_estimators=150)
best_estimators = 150  # Using 150 as the best value
rf_model_best = RandomForestClassifier(n_estimators=best_estimators, random_state=42)
rf_model_best.fit(x_train, y_train)
best_rf_predictions = rf_model_best.predict(x_test)

cm_best_rf = confusion_matrix(y_test, best_rf_predictions)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_best_rf, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title(f'Confusion Matrix Heatmap (Random Forest with {best_estimators} Estimators)')
plt.show()
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns

# Instantiate and train the Linear Regression model
lin_reg_model = LinearRegression()
lin_reg_model.fit(x_train, y_train)

# Predictions on the test set (continuous output)
lin_reg_predictions = lin_reg_model.predict(x_test)

# Convert predictions to binary classification
lin_reg_predictions_binary = (lin_reg_predictions >= 0.5).astype(int)

# Accuracy
lin_reg_accuracy = accuracy_score(y_test, lin_reg_predictions_binary)
print("Accuracy of Linear Regression model is ", lin_reg_accuracy * 100, "%")
print("Classification Report of Linear Regression model is \n", classification_report(y_test, lin_reg_predictions_binary))

# Confusion Matrix
cm_lin_reg = confusion_matrix(y_test, lin_reg_predictions_binary)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_lin_reg, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix Heatmap (Linear Regression)')
plt.show()

# ROC Curve (for binary classification)
if len(le.classes_) == 2:  # Only applicable for binary classification
    y_prob_lin_reg = lin_reg_predictions  # Continuous outputs from Linear Regression
    fpr_lin_reg, tpr_lin_reg, _ = roc_curve(y_test, y_prob_lin_reg)
    auc_score_lin_reg = roc_auc_score(y_test, y_prob_lin_reg)

    plt.figure(figsize=(8, 6))
    plt.plot(fpr_lin_reg, tpr_lin_reg, color='purple', label=f'AUC = {auc_score_lin_reg:.2f}')
    plt.plot([0, 1], [0, 1], color='red', linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve (Linear Regression)')
    plt.legend()
    plt.show()

# Feature Correlation Matrix Heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(encoded_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix Heatmap')
plt.show()

# Linear Regression with Regularization (Ridge and Lasso)
regularization_values = [0.1, 1, 10, 100, 1000]  # Different values for the regularization parameter alpha

# Ridge Regression (L2 regularization)
ridge_accuracies = []
for alpha_value in regularization_values:
    ridge_model = Ridge(alpha=alpha_value)
    ridge_model.fit(x_train, y_train)
    
    # Predictions and accuracy for Ridge
    ridge_predictions = ridge_model.predict(x_test)
    ridge_predictions_binary = (ridge_predictions >= 0.5).astype(int)
    ridge_acc = accuracy_score(y_test, ridge_predictions_binary)
    ridge_accuracies.append(ridge_acc)

# Lasso Regression (L1 regularization)
lasso_accuracies = []
for alpha_value in regularization_values:
    lasso_model = Lasso(alpha=alpha_value)
    lasso_model.fit(x_train, y_train)
    
    # Predictions and accuracy for Lasso
    lasso_predictions = lasso_model.predict(x_test)
    lasso_predictions_binary = (lasso_predictions >= 0.5).astype(int)
    lasso_acc = accuracy_score(y_test, lasso_predictions_binary)
    lasso_accuracies.append(lasso_acc)

# Plot Effect of Regularization (Ridge and Lasso) on Accuracy
plt.figure(figsize=(8, 6))
plt.plot(regularization_values, ridge_accuracies, marker='o', color='blue', label='Ridge Regression Accuracy')
plt.plot(regularization_values, lasso_accuracies, marker='o', color='orange', label='Lasso Regression Accuracy')
plt.xscale('log')  # Log scale for better visualization
plt.xlabel('Regularization Strength (alpha)')
plt.ylabel('Accuracy')
plt.title('Effect of Regularization on Accuracy (Ridge vs Lasso)')
plt.legend()
plt.grid(True)
plt.show()

# Confusion Matrix for Ridge Regression with Best Alpha (alpha=1)
best_alpha_ridge = 1
ridge_model_best = Ridge(alpha=best_alpha_ridge)
ridge_model_best.fit(x_train, y_train)
ridge_predictions_best = ridge_model_best.predict(x_test)
ridge_predictions_best_binary = (ridge_predictions_best >= 0.5).astype(int)

cm_ridge_best = confusion_matrix(y_test, ridge_predictions_best_binary)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_ridge_best, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title(f'Confusion Matrix Heatmap (Ridge Regression with alpha={best_alpha_ridge})')
plt.show()

# Confusion Matrix for Lasso Regression with Best Alpha (alpha=1)
best_alpha_lasso = 1
lasso_model_best = Lasso(alpha=best_alpha_lasso)
lasso_model_best.fit(x_train, y_train)
lasso_predictions_best = lasso_model_best.predict(x_test)
lasso_predictions_best_binary = (lasso_predictions_best >= 0.5).astype(int)

cm_lasso_best = confusion_matrix(y_test, lasso_predictions_best_binary)
plt.figure(figsize=(8, 6))
sns.heatmap(cm_lasso_best, annot=True, fmt='d', cmap='Blues', xticklabels=le.classes_, yticklabels=le.classes_)
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title(f'Confusion Matrix Heatmap (Lasso Regression with alpha={best_alpha_lasso})')
plt.show()
```
