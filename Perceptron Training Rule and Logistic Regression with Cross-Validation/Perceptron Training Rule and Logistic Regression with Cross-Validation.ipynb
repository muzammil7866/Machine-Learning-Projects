# Cross Validation using K fold = 5 (Perceptron Training Rule)

```python
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import Perceptron
from sklearn.metrics import classification_report, confusion_matrix

df = pd.read_csv("cancer.csv")
print(df.shape)

x = df.drop(columns=['Id', 'Diagnosis'])
y = df['Diagnosis']

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.3, random_state=42)

print(x_train.shape)

kf = KFold(n_splits=5, shuffle=True, random_state=42)
model = Perceptron()

metrics = {
    'Training': {'Accuracy_M': [], 'Accuracy_B': [], 'Precision_M': [], 'Precision_B': [], 'Recall_M': [], 'Recall_B': [], 'F1_M': [], 'F1_B': []},
    'Validation': {'Accuracy_M': [], 'Accuracy_B': [], 'Precision_M': [], 'Precision_B': [], 'Recall_M': [], 'Recall_B': [], 'F1_M': [], 'F1_B': []}
}

def class_accuracy(cm, class_index):
    if class_index == 1:  
        TP = cm[class_index, class_index]  
        FP = cm[0, class_index]  
        return TP / (TP + FP) if (TP + FP) != 0 else 0
    else:  
        TN = cm[class_index, class_index]
        FN = cm[1, class_index]  
        return TN / (TN + FN) if (TN + FN) != 0 else 0

for train_index, val_index in kf.split(x_train):
    print(train_index.shape)
    print(val_index.shape)
    x_ktrain, x_val = x_train.iloc[train_index], x_train.iloc[val_index]
    y_ktrain, y_val = y_train.iloc[train_index], y_train.iloc[val_index]

    model.fit(x_ktrain, y_ktrain)

    y_pred_train = model.predict(x_ktrain)
    y_pred_val = model.predict(x_val)

    train_cm = confusion_matrix(y_ktrain, y_pred_train)
    accuracy_M_train = class_accuracy(train_cm, 1)  # Accuracy for Malignant (Class M)
    accuracy_B_train = class_accuracy(train_cm, 0)  # Accuracy for Benign (Class B)

    train_report = classification_report(y_ktrain, y_pred_train, output_dict=True)
    
    metrics['Training']['Accuracy_M'].append(accuracy_M_train)
    metrics['Training']['Accuracy_B'].append(accuracy_B_train)
    metrics['Training']['Precision_M'].append(train_report['M']['precision'])
    metrics['Training']['Precision_B'].append(train_report['B']['precision'])
    metrics['Training']['Recall_M'].append(train_report['M']['recall'])
    metrics['Training']['Recall_B'].append(train_report['B']['recall'])
    metrics['Training']['F1_M'].append(train_report['M']['f1-score'])
    metrics['Training']['F1_B'].append(train_report['B']['f1-score'])

    val_cm = confusion_matrix(y_val, y_pred_val)
    accuracy_M_val = class_accuracy(val_cm, 1)  # Accuracy for Malignant (Class M)
    accuracy_B_val = class_accuracy(val_cm, 0)  # Accuracy for Benign (Class B)

    val_report = classification_report(y_val, y_pred_val, output_dict=True)
    metrics['Validation']['Accuracy_M'].append(accuracy_M_val)
    metrics['Validation']['Accuracy_B'].append(accuracy_B_val)
    metrics['Validation']['Precision_M'].append(val_report['M']['precision'])
    metrics['Validation']['Precision_B'].append(val_report['B']['precision'])
    metrics['Validation']['Recall_M'].append(val_report['M']['recall'])
    metrics['Validation']['Recall_B'].append(val_report['B']['recall'])
    metrics['Validation']['F1_M'].append(val_report['M']['f1-score'])
    metrics['Validation']['F1_B'].append(val_report['B']['f1-score'])

model.fit(x_train, y_train)
y_pred_test = model.predict(x_test)

test_cm = confusion_matrix(y_test, y_pred_test)
accuracy_M_test = class_accuracy(test_cm, 1)  # Accuracy for Malignant (Class M)
accuracy_B_test = class_accuracy(test_cm, 0)  # Accuracy for Benign (Class B)

test_report = classification_report(y_test, y_pred_test, output_dict=True)

test_metrics = {
    'Accuracy_M': accuracy_M_test,
    'Accuracy_B': accuracy_B_test,
    'Precision_M': test_report['M']['precision'],
    'Precision_B': test_report['B']['precision'],
    'Recall_M': test_report['M']['recall'],
    'Recall_B': test_report['B']['recall'],
    'F1_M': test_report['M']['f1-score'],
    'F1_B': test_report['B']['f1-score']
}

print("Training Metrics (Averaged Across Folds):")
for metric in metrics['Training']:
    print(f"{metric}: {sum(metrics['Training'][metric]) / len(metrics['Training'][metric]):.2f}")

print("\nValidation Metrics (Averaged Across Folds):")
for metric in metrics['Validation']:
    print(f"{metric}: {sum(metrics['Validation'][metric]) / len(metrics['Validation'][metric]):.2f}")

print("\nTest Metrics:")
for metric in test_metrics:
    print(f"{metric}: {test_metrics[metric]:.2f}")
```

# Cross Validation using K fold = 5 (Logistic Regression)

```python
import pandas as pd
from sklearn.model_selection import train_test_split, KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

df = pd.read_csv("cancer.csv")

x = df.drop(columns=['Id', 'Diagnosis'])
y = df['Diagnosis']

x_train, x_test, y_train, y_test = train_test_split(x, y, stratify=y, test_size=0.3, random_state=42)

kf = KFold(n_splits=5, shuffle=True, random_state=42)
model = LogisticRegression(max_iter = 2200)

metrics = {
    'Training': {'Accuracy_M': [], 'Accuracy_B': [], 'Precision_M': [], 'Precision_B': [], 'Recall_M': [], 'Recall_B': [], 'F1_M': [], 'F1_B': []},
    'Validation': {'Accuracy_M': [], 'Accuracy_B': [], 'Precision_M': [], 'Precision_B': [], 'Recall_M': [], 'Recall_B': [], 'F1_M': [], 'F1_B': []}
}

def class_accuracy(cm, class_index):
    if class_index == 1:  # For positive class (Malignant)
        TP = cm[class_index, class_index]  
        FP = cm[0, class_index]  # False positives for the class
        return TP / (TP + FP) if (TP + FP) != 0 else 0
    else:  # For negative class (Benign)
        TN = cm[class_index, class_index]
        FN = cm[1, class_index]  # False negatives for the class
        return TN / (TN + FN) if (TN + FN) != 0 else 0

for train_index, val_index in kf.split(x_train):
    x_ktrain, x_val = x_train.iloc[train_index], x_train.iloc[val_index]
    y_ktrain, y_val = y_train.iloc[train_index], y_train.iloc[val_index]

    model.fit(x_ktrain, y_ktrain)

    y_pred_train = model.predict(x_ktrain)
    y_pred_val = model.predict(x_val)

    train_cm = confusion_matrix(y_ktrain, y_pred_train)
    accuracy_M_train = class_accuracy(train_cm, 1)  # Accuracy for Malignant (Class M)
    accuracy_B_train = class_accuracy(train_cm, 0)  # Accuracy for Benign (Class B)

    train_report = classification_report(y_ktrain, y_pred_train, output_dict=True)
    
    metrics['Training']['Accuracy_M'].append(accuracy_M_train)
    metrics['Training']['Accuracy_B'].append(accuracy_B_train)
    metrics['Training']['Precision_M'].append(train_report['M']['precision'])
    metrics['Training']['Precision_B'].append(train_report['B']['precision'])
    metrics['Training']['Recall_M'].append(train_report['M']['recall'])
    metrics['Training']['Recall_B'].append(train_report['B']['recall'])
    metrics['Training']['F1_M'].append(train_report['M']['f1-score'])
    metrics['Training']['F1_B'].append(train_report['B']['f1-score'])

    val_cm = confusion_matrix(y_val, y_pred_val)
    accuracy_M_val = class_accuracy(val_cm, 1)  # Accuracy for Malignant (Class M)
    accuracy_B_val = class_accuracy(val_cm, 0)  # Accuracy for Benign (Class B)

    val_report = classification_report(y_val, y_pred_val, output_dict=True)
    metrics['Validation']['Accuracy_M'].append(accuracy_M_val)
    metrics['Validation']['Accuracy_B'].append(accuracy_B_val)
    metrics['Validation']['Precision_M'].append(val_report['M']['precision'])
    metrics['Validation']['Precision_B'].append(val_report['B']['precision'])
    metrics['Validation']['Recall_M'].append(val_report['M']['recall'])
    metrics['Validation']['Recall_B'].append(val_report['B']['recall'])
    metrics['Validation']['F1_M'].append(val_report['M']['f1-score'])
    metrics['Validation']['F1_B'].append(val_report['B']['f1-score'])

model.fit(x_train, y_train)
y_pred_test = model.predict(x_test)

test_cm = confusion_matrix(y_test, y_pred_test)
accuracy_M_test = class_accuracy(test_cm, 1)  # Accuracy for Malignant (Class M)
accuracy_B_test = class_accuracy(test_cm, 0)  # Accuracy for Benign (Class B)

test_report = classification_report(y_test, y_pred_test, output_dict=True)

test_metrics = {
    'Accuracy_M': accuracy_M_test,
    'Accuracy_B': accuracy_B_test,
    'Precision_M': test_report['M']['precision'],
    'Precision_B': test_report['B']['precision'],
    'Recall_M': test_report['M']['recall'],
    'Recall_B': test_report['B']['recall'],
    'F1_M': test_report['M']['f1-score'],
    'F1_B': test_report['B']['f1-score']
}

print("Training Metrics (Averaged Across Folds):")
for metric in metrics['Training']:
    print(f"{metric}: {sum(metrics['Training'][metric]) / len(metrics['Training'][metric]):.2f}")

print("\nValidation Metrics (Averaged Across Folds):")
for metric in metrics['Validation']:
    print(f"{metric}: {sum(metrics['Validation'][metric]) / len(metrics['Validation'][metric]):.2f}")

print("\nTest Metrics:")
for metric in test_metrics:
    print(f"{metric}: {test_metrics[metric]:.2f}")
```
