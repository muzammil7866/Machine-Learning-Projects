```python
#QUESTION 1 NAIVE BAYES WITH SCRATCH

import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score

data_train = {
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny', 'Sunny', 'Rain'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal'],
    'Wind': ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Weak', 'Weak'],
    'PlayTennis': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes']
}

df_train = pd.DataFrame(data_train)


X_train = df_train[['Outlook', 'Temperature', 'Humidity', 'Wind']]
Y_train = df_train[['PlayTennis']]

data_test = {
    'Outlook': ['Sunny', 'Overcast', 'Overcast', 'Rain'],
    'Temperature': ['Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['Normal', 'High', 'Normal', 'High'],
    'Wind': ['Strong', 'Strong', 'Weak', 'Strong'],
    'PlayTennis': ['Yes', 'Yes', 'Yes', 'No']
}

df_test = pd.DataFrame(data_test)

X_test = df_test[['Outlook', 'Temperature', 'Humidity', 'Wind']]
Y_test = df_test[['PlayTennis']]

dataset = {}
for i in X_train.columns:
    dataset[i] = {}
    values = []
    for j in np.unique(df_train[i]):
        values.append(j)
    #print(values)
    #print(i, "\n")
    for j in values:
        dataset[i][j] = {}
        yes = df_train[i][(df_train[i] == j) & (df_train['PlayTennis'] == 'Yes')]
        no = df_train[i][(df_train[i] == j) & (df_train['PlayTennis'] == 'No')]
        dataset[i][j]['Yes'] = yes.count()/(yes.count() + no.count())
        dataset[i][j]['No'] = no.count()/(yes.count() + no.count())
        
        #print(f"{j} yes", yes.count())
        #print(f"{j} no", no.count())
        #print("-")

print(dataset)

yesy = Y_train['PlayTennis'][Y_train['PlayTennis'] == 'Yes']
noy = Y_train['PlayTennis'][Y_train['PlayTennis'] == 'No']

yesy = yesy.shape[0]/Y_train.shape[0]
noy = noy.shape[0]/Y_train.shape[0]
#print(yesy)
#print(noy)



test1y = yesy*dataset['Outlook']['Sunny']['Yes']*dataset['Temperature']['Mild']['Yes']*dataset['Humidity']['Normal']['Yes']*dataset['Wind']['Strong']['Yes']
#print(test1y)
test1n = noy*dataset['Outlook']['Sunny']['No']*dataset['Temperature']['Mild']['No']*dataset['Humidity']['Normal']['No']*dataset['Wind']['Strong']['No']
#print(test1n)

final1 = test1y/(test1y+test1n)
#print(final1)


test2y = yesy*dataset['Outlook']['Overcast']['Yes']*dataset['Temperature']['Mild']['Yes']*dataset['Humidity']['High']['Yes']*dataset['Wind']['Strong']['Yes']
#print(test2y)
test2n = noy*dataset['Outlook']['Overcast']['No']*dataset['Temperature']['Mild']['No']*dataset['Humidity']['High']['No']*dataset['Wind']['Strong']['No']
#print(test2n)

final2 = test2y/(test2y+test2n)
#print(final2)

test3y = yesy*dataset['Outlook']['Overcast']['Yes']*dataset['Temperature']['Hot']['Yes']*dataset['Humidity']['Normal']['Yes']*dataset['Wind']['Weak']['Yes']
#print(test3y)
test3n = noy*dataset['Outlook']['Overcast']['No']*dataset['Temperature']['Hot']['No']*dataset['Humidity']['Normal']['No']*dataset['Wind']['Weak']['No']
#print(test3n)

final3 = test3y/(test3y+test3n)
#print(final3)

test4y = yesy*dataset['Outlook']['Rain']['Yes']*dataset['Temperature']['Mild']['Yes']*dataset['Humidity']['High']['Yes']*dataset['Wind']['Strong']['Yes']
#print(test4y)
test4n = noy*dataset['Outlook']['Rain']['No']*dataset['Temperature']['Mild']['No']*dataset['Humidity']['High']['No']*dataset['Wind']['Strong']['No']
#print(test4n)

final4 = test4y/(test4y+test4n)
#print(final4)

probabilities = [final1,final2, final3, final4]

print("\n", probabilities)
finalPredictions = []
for i in range(len(probabilities)):
    if(probabilities[i] >= 0.5):
        finalPredictions.append('Yes')
    else:
        finalPredictions.append('No')
print("\nPredicted Values: ", finalPredictions, "\n")
print("True values:", Y_test.values)

accuracy = accuracy_score(Y_test.values, finalPredictions)
print(f"ACCURACY:  {accuracy*100}%")




```

```python
#QUESTION 1 NAIVE BAYES BUILT-IN

import pandas as pd
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import accuracy_score
from sklearn import preprocessing

data_train = {
    'Outlook': ['Sunny', 'Sunny', 'Overcast', 'Rain', 'Rain', 'Rain', 'Overcast', 'Sunny', 'Sunny', 'Rain'],
    'Temperature': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal'],
    'Wind': ['Weak', 'Strong', 'Weak', 'Weak', 'Weak', 'Strong', 'Strong', 'Weak', 'Weak', 'Weak'],
    'PlayTennis': ['No', 'No', 'Yes', 'Yes', 'Yes', 'No', 'Yes', 'No', 'Yes', 'Yes']
}

df_train = pd.DataFrame(data_train)

le = preprocessing.LabelEncoder()

df_train = df_train.apply(le.fit_transform)

X_train = df_train[['Outlook', 'Temperature', 'Humidity', 'Wind']]
Y_train = df_train['PlayTennis']

data_test = {
    'Outlook': ['Sunny', 'Overcast', 'Overcast', 'Rain'],
    'Temperature': ['Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['Normal', 'High', 'Normal', 'High'],
    'Wind': ['Strong', 'Strong', 'Weak', 'Strong'],
    'PlayTennis': ['Yes', 'Yes', 'Yes', 'No']
}

df_test = pd.DataFrame(data_test)

df_test = df_test.apply(le.fit_transform)

X_test = df_test[['Outlook', 'Temperature', 'Humidity', 'Wind']]
Y_test = df_test['PlayTennis']


#print(df_train)
#print(df_test)

nb_model = CategoricalNB()

nb_model.fit(X_train, Y_train)

predictions = nb_model.predict(X_test)

accuracy = accuracy_score(Y_test, predictions)
print(accuracy*100, "%")

print("Predicted values:", predictions)
print("True values:", Y_test.values)



```
