# 1) Reading the Dataset and Displaying

```python
import pandas as pd

df = pd.read_csv("../Data/Multiple Disease Raw Data.csv")
print(df.info())
print(df.describe())
df.head(3)

print(df['KidneyDisease'].value_counts())

```

# 2) Performing Feature Engineering to remove the unnecessary columns (irrlevant targets), taking KidneyDisease as Target!

```python
print(df.columns)
print(df.head(3))
df.drop(columns = ['Asthma', 'SkinCancer'], inplace = True)
print("\n\n\nValue Splits in KidneyDisease: ", df['KidneyDisease'].value_counts())
```

# 3) Data Preprocessing (Checking and Handling for Null Values)

```python
from sklearn.preprocessing import StandardScaler

print(df.head(3))
print(df.isnull().sum())

print(df['MentalHealth'].unique())

df['MentalHealth'].fillna(df['MentalHealth'].mode()[0], inplace=True)
df['GenHealth'].fillna(df['GenHealth'].mode()[0], inplace=True)

print(df.isnull().sum())
```

# 4) Data Standardization to normalise data (Using StandardScaler that uses the Z-score)

# 5) Encoding Categorical columns using LabelEncoder()

```python
print(df.info())

bmi = df[['BMI']]
PhysicalHealth = df[['PhysicalHealth']]
MentalHealth = df[['MentalHealth']]
SleepTime = df[['SleepTime']]

scaler = StandardScaler()

bmi = scaler.fit_transform(bmi)
PhysicalHealth = scaler.fit_transform(PhysicalHealth)
MentalHealth = scaler.fit_transform(MentalHealth)
SleepTime = scaler.fit_transform(SleepTime)

df['BMI'] = bmi
df['PhysicalHealth'] = PhysicalHealth
df['MentalHealth'] = MentalHealth
df['SleepTime'] = SleepTime

df.head(3)
```

# 6)  Seperating Features and Target

```python
x = df.copy()
x.drop(columns = 'KidneyDisease', inplace = True)

x = pd.get_dummies(x, drop_first=True)

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
encoded_df = df.apply(le.fit_transform)

y = encoded_df[['KidneyDisease']]

print(x)
y
```

# 7) Applying Test Train Split for 80% on training and 20% on testing

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y,stratify = y, random_state = 42, test_size = 0.2)

print("X Train: ",x_train.shape[0], "rows with ", x_train.shape[1], "features")
print("X Test: ", x_test.shape[0], "rows", x_test.shape[1], "features")
print("Y Train: ", y_train.shape[0], "rows", y_train.shape[1], "feature")
print("Y Test: ", y_test.shape[0], "rows", y_test.shape[1], "feature")

```

# 8) Training the Model with Linear Regression and generating Predictions

```python
from sklearn.linear_model import LinearRegression, LogisticRegression

liR = LinearRegression()
liR.fit(x_train, y_train)

predictions = liR.predict(x_test)

print("Prediction: ", predictions)
```

# 9) Determining Final Classes

```python
final = []
for i in predictions:
    if(i >= 0.5):
        final.append(1)
    else:
        final.append(0)


finalPredictions = pd.DataFrame({'Target': y_test['KidneyDisease'][:].values, "Our Result": final})

print(finalPredictions['Target'].value_counts())
print(finalPredictions['Our Result'].value_counts())
finalPredictions
```

# 10) Evaluating model using different metrices

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report,confusion_matrix

accuracy = accuracy_score(final, y_test)

print("Accuracy: %.2f %%" % (accuracy * 100))
print("\nClassification Report:")
print(classification_report(y_test, final, zero_division = 1))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, final))


```
