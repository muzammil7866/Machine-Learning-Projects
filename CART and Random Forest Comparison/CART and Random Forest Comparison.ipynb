# QUESTION 1 CART

```python
#QUESTION 1 CART
import pandas as pd
#traindata
trainData = {'x1': [60,62, 67, 70, 71, 72, 75, 78], 'x2': [22,25,24,20,15,14,14,11], 'y': [0,0,0,0,1,1,1,1]}

df = pd.DataFrame(trainData)
print(df)

X = df[['x1', 'x2']]
Y = df['y']

print(X)
print(Y)


```

```python
#test data
testData = {
    'x1': [61, 71, 73, 79],
    'x2': [23, 19, 15, 13],
    'y': [0, 0, 1, 1]
}

df2 = pd.DataFrame(testData)

X2 = df2[['x1', 'x2']]
Y2 = df2['y']

# Display the DataFrame
print(df2)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=42)

#training the model
clf.fit(X,Y)

y_pred = clf.predict(X2)
print(y_pred)
print(Y2)



```

## Why random_state = 42?
### The number 42 is often used humorously as a reference to The Hitchhiker's Guide to the Galaxy by Douglas Adams, where 42 is given as "the answer to the ultimate question of life, the universe, and everything." It's become somewhat of a convention in data science and programming communities.

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y2, y_pred)
print(accuracy*100, "%")
```

# QUESTION 1 RANDOM FOREST

```python
#QUESTION 1 PART 2 RANDOM FORST

from sklearn.ensemble import RandomForestClassifier

#n_estimator is the number of classifiers used in the random forest

rf_model = RandomForestClassifier(n_estimators=100, random_state=42)

rf_model.fit(X, Y)

predictions = rf_model.predict(X2)

accuracy3 = accuracy_score(Y2, predictions)
print(accuracy3*100, "%")

# Step 9: Print test set predictions and true labels
print("Predicted labels:", predictions)
print("True labels:", Y2.values)



```

# Comparison b/w decision tree and random forest for our dataset

### The observed results i.e. accuracy score from decision tree classifier and random forests are 75% and 100% respectively which means that random forest performed better for our dataset in this case. 

## Possible reason:

### This could be because it is seen that decision trees happen to overfit data which in the case of random forest, it performs better. Another reason could be that random forest combines predictions from multiple classifiers which reduces individual error and in turn leads to better accuracy.

# QUESTION 2 RANDOM FOREST

```python
#QUESTION 2 CART
trainData = {
    'x1': [60, 62, 67, 70, 71],
    'x2': [22, 25, 24, 20, 15],
    'y': [140, 155, 159, 179, 192]
}

df3 = pd.DataFrame(trainData)

print(df3)

X3 = df3[['x1', 'x2']]
Y3 = df3['y']
```

```python
testData = {
    'x1': [72, 75, 78],
    'x2': [14, 14, 11],
    'y': [200, 212, 215]
}

df4 = pd.DataFrame(testData)

print(df4)

X4 = df4[['x1', 'x2']]
Y4 = df4['y']
```

```python
from sklearn.tree import DecisionTreeRegressor



reg = DecisionTreeRegressor(random_state=42)

#training the model
reg.fit(X3,Y3)

y_pred2 = reg.predict(X4)
print(y_pred2)
print(Y4)



```

```python
#optional part

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(Y4, y_pred2)
r2 = r2_score(Y4, y_pred2)

print(f'Mean Squared Error: {mse}')
print(f'R² Score: {r2}')
```

# QUESTION 2 DECISION TREE

```python
#QUESTION 2 PART 2 RANDOM FORST
from sklearn.ensemble import RandomForestRegressor

rf_model2 = RandomForestRegressor(n_estimators=100, random_state=42)

rf_model2.fit(X3, Y3)

predictions2 = rf_model2.predict(X4)

mse2 = mean_squared_error(Y4, predictions2)
r22 = r2_score(Y4, predictions2)

print(f'Mean Squared Error: {mse2}')
print(f'R² Score: {r22}')

print("Predicted values:", predictions2)
print("True values:", Y4.values)

```
