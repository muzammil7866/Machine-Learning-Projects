```python
import numpy as np

 

attributes = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])

labels = np.array([0, 1, 1, 1])

 

weights = np.array([0.9, 0.9])

threshold = 0.5

learning_rate = 0.1

epochs = 100

 

for epoch in range(epochs):

    for input_data, label in zip(attributes, labels):

        activation = np.dot(input_data, weights)

        prediction = 1 if activation >= threshold else 0

        error = label - prediction

        weights += learning_rate * error * input_data

 

print("Trained Weights:", weights)

 

for input_data, label in zip(attributes, labels):

    activation = np.dot(input_data, weights)

    prediction = 1 if activation >= threshold else 0

    print(f"Input: {input_data}, Actual: {label}, Predicted: {prediction}")
```
