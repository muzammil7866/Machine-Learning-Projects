# QUESTION 1 (A)

```python
import pandas as pd

data = {'D': [0.009, 0.001], '-D': [0.099,0.891]}
df = pd.DataFrame(data,  index = ['+', '-'])
print(df)


print(f"PROBABILITY OF +VE GIVEN DISEASE: {df['D']['+']}")



probPosGivenDisease = df['D']['+']
probDis = df['D'].sum()
probPos = probDis*probPosGivenDisease + (1-probDis) * (df['-D']['+'])
print(f"PROBABILITY OF DISEASE GIVEN +VE: {(probPosGivenDisease*probDis)/probPos}")
```

# QUESTION 1 (B)

```python


p_spam = 0.7        
p_low_priority = 0.2  
p_high_priority = 0.1 


p_free_given_spam = 0.8  
p_free_given_low = 0.3 
p_free_given_high = 0.1  

p_free = (p_free_given_spam * p_spam) + (p_free_given_low * p_low_priority) + (p_free_given_high * p_high_priority)

p_spam_given_free = (p_free_given_spam * p_spam) / p_free

print(f"\nProbability that the email is spam given that it contains the word 'free': P(Spam | 'free') = {p_spam_given_free:.4f}")
```
